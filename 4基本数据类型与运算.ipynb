## 1 Java提供的基本数据类型：
    Java语言一共提供了8种原始的数据类型：byte，short, int, long, float, double, char, boolean. 这种基本类型的数据变量在声明之后立刻在栈上分配内存空间
    原始数据类型在传递参数时是值传递，封装类型是按引用传递的
    
    除了基本类型外，Java基本都是引用类型
    
    int和Integer有什么区别：int时原始数据类型，Integer是Java语言为int提供的封装类，Java为每个原始类型都提供了封装类
    

## 2 什么是不可变类：
    一个对象一旦被创建出来，在其整个生命周期中，她的成员变量就不能被修改了，类似于const，只允许别的程序都，不允许别的程序修改
    创建不可变类的基本原则：
        1 类中所有成员变量被private所修饰
        2 类中没有写或者修改成员变量的方法
        3 确保类中所有方法不会被子类覆盖
        4 如果一个类成员不是不可变量，那么在成员初始化或者使用get方法获取该成员变量时，需要通过clone方法来确保类的不可变性
        5 如果有必要，可使用object类的equals方法和hashcode()方法。

## 3 值传递和引用传递有哪些区别：
    1 值传递：
        实参将它的值传递给形参，形参只是用实参的值初始化一个临时的存储单元，因此形参与实参虽然有相同的值，但是有着不同的存储单元，因此，形参的改变不会影响实参的值
    2 引用传递：
        方法调用时，传递的是对象（可以看作是对象的地址），这时形参与实参的对象指向同一块存储单元，因此，对形参的修改就是对实参的修改（修改内容不修改地址）。

## 9 ++i 与 i++区别：
    不同点在于i++是在程序执行完毕后进行自增，而++i是在程序执行前进行自增

## 10 如何实现无符号的右移：
    两种右移操作：">>" 和 ">>>",">>"有符号右移，">>>"无符号右移
    
    "<<"运算符表示左移，左移n位表示原来的值乘2的n次方
    
    为了保证移位的有效性，以使移动的位数不超过32bit，采用取余的方法，a>>n等价于a>>(n%32)

## 11 char型变量中是否可以存储一个中文：
    Java中，默认使用Unicode编码方式，即每个字符占用两个字节，因此可以用来存储中文
