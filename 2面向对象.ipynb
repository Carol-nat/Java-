## 1 面向对象和面向过程有什么区别：
    1 出发点不同（面向对象强调：将问题域的要领直接映射到对象及对象的接口上。面上过程：过程的抽象化和模块化）
    2 层次逻辑关系不同（面向对象是用类的层次结构来体现类之间的继承与发展，面向过程是使用模块的层次结构概括模块或模块的关系与功能）
    3 数据处理方式与控制程序方式不同
    4 分析设计与编码转换方式不同

## 2 面向对象的特征：
    1 抽象
    2 继承（子类可以修改或增加新的方法使之更适合特殊的需要）
    3 封装 （将客观事物抽象成类）
    4 多态 （允许不同类的对象对同一消息做出响应，分为参数化多态和包含多态）

## 3 面向对象的开发方式：
    1 较高的开发效率
    2 保证软件的鲁棒性
    3 保证软件的高可维护性

## 4 什么是继承：
    特性：
    1 Java语言不支持多重继承，单继承，多实现
    2 子类只能继承父类的非私有（public，protected）成员变量与方法
    3 子类定义的成员变量与父类中定义的成员变量重名时，子类中的成员变量复写父类的成员变量
    4 子类方法与父类方法有相同的函数签名时，覆盖

## 5 组合与继承有什么区别：
    组合与继承是面向对象中两种代码复用的方式
    组合：在新类里面创建原有类的对象，重复利用已有类的功能（has-a的关系）
    继承：根据其他类的实现来定义一个类的实现（is-a的关系）

## 6 多态的实现机制：
    1 方法的重载（同一个类有多种同名但有多种不同参数的方法），编译时多态
    2 方法的覆盖（子类方法覆盖父类方法，同样方法在子类和父类中有着不同的表现形式），运行时多态

## 7 重载（overload）与覆盖（override）有什么区别：
    
   ### 重载：
    1 重载是通过不同的方法参数来区分，（不同的参数个数，不同的参数类型，不同的参数顺序）
    2 重载不能通过方法的访问权限，返回值类型和抛出的异常来进行重载
    3 对于继承来讲，如果基类方法的访问权限是private，那么就不能在子类中进行重载
    
   ### 覆盖
    1 子类中的覆盖方法和父类中的被覆盖方法有着相同的函数名和参数
    2 子类中的覆盖方法的返回值和父类中被覆盖方法的返回值一样
    3 抛出的异常也一样
    4 基类中被覆盖的方法不能是private
    
    区别：
    1 覆盖是子类与父类之间的关系，垂直关系，重载是同一个类中方法之间的关系，水平关系
    2 覆盖只能由一个方法或者一对方法产生关系；重载是多个方法之间的关系
    3 覆盖要求参数列表相同，重载要求参数列表不同
    4 覆盖关系中，调用方法体是根据对象的类型来决定，而重载是更居调用时的实参列表与形参表来选择方法体。
    
    

## 8 抽象类 abstract class和接口 interface有什么异同
    抽象类和接口都是支持抽象类定义的两种机制，前者是一个实体，后者是一个概念，
    相同点：
        1 都不能被实例化
        2 接口的实现类或抽象类的子类都只有实现了接口或抽象类中的方法后才能被实例化。
    不同点：
        1 接口只有定义，其方法不能再接口中实现，只有实现接口的类才能实现接口中定义的方法，而抽象类其方法可以在抽象类中实现。
        2 接口需要实现（implements），但抽象类只能被继承
        3 接口强调特定功能的实现，而抽象类强调所属关系
        4 接口中定义的变量默认为 public static final，成员方法都是被public abstract修饰；抽象类可以有自己的数据成员变量，也可以有非抽象的成员方法，成员变量默认为default，当然也可定义为private，protected，public，这些变量可以在子类中被重新定义或赋值。抽象类中的抽象方法不能用private，static，synchronized，native等修饰符修饰，同时方法必须以分号结尾，不带花括号
        5 接口被运用于实现比较常用的功能，抽象类更倾向于充当公共类的角色。

## 9 内部类有哪些：
    四种：静态内部类（非静态内部类中不能有静态成员），成员内部类，局部内部类，匿名内部类
    匿名内部类是没有类名的内部类，不适用关键字class，extends，implement，没有构造函数，必须继承其他类或实现其他接口。使用原则：
    1 不能有构造函数
    2 不能定义静态变量，方法，类
    3 不能用public protected private， static
    4 只能创建匿名内部类的一个实例
    5 一个匿名内部类一定实在new的后面，必须继承一个父类或者实现一个接口
    6 匿名内部类时局部内部类

## 11 this 和 super有什么区别：
    this用来指向当前实例对象，一个重要的作用就是用来区分对象的成员变量与方法的形参
    super用来访问父类的方法或成员变量
